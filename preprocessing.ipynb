The Stationary Wavelet Transform (SWT) is an extension of the Discrete Wavelet Transform (DWT) designed to overcome the lack of translation-invariance of the discrete wavelet transform. SWT is feasible for feature extraction, change detection and pattern recognition due to shift-invariant and non-decimated properties. In SWT, after the signal is convolved with high and low pass filters, no decimation is performed, thereby the number of obtained coefficients cA and cD at each decomposition level is the same as the number of samples in the original signal. Frequently used in EMG and ECG, its application will be analysed for financial time series.

3 wavelets will be considered - Meyer, Debauchies & Haar. Since we are analysing less smooth data like stock returns, less smooth wavelet functions like 3 mentioned are the most suitable for analysis

Five-level SWT decomposition of an input signal will be conducted.

```python
def swt_transform(df, col_names):
    wavelet = 'dmey'
    level = 5
    col_names_all= [(x + '_{}{}'.format(j, i)) for x in col_names for i in range(1, 6)
            for j in ['A', 'D']]
    features = list()
        
    swt = df.apply(lambda col:
             pywt.swt(col, 
             wavelet = 'dmey', level= 5,
             start_level=0 , axis=0, 
             trim_approx=False, norm=False), axis = 0)
    
    for i in itertools.chain.from_iterable(swt):
        for j in i:
            features.append(j)
    d = dict(zip(col_names_all, features))
    return pd.DataFrame(d)


X_train_swt = swt_transform(X_train, col_names = ['tsla', 'covid', 'ixic', 'gspc', 'djia'])
X_test_swt = swt_transform(X_test, col_names = ['tsla', 'covid', 'ixic', 'gspc', 'djia'] )
```

Different combinations of features will be tested. They are namely:
    1. Level 5 cA and CD for all features
    2. Level 5 cA for all features except COVID-19, CD for all features
    3. PCA 

```python
# PCA

# Create the scaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train_swt)
X_test_std = scaler.transform(X_test_swt)

# Create the PCA instance and fit and transform the data with pca
pca = PCA()
pc = pca.fit(X_train_std)

print(pc.explained_variance_ratio_)
print(pc.explained_variance_ratio_.cumsum()) #13 For 90% of variance

n_components =13
pca_n2 = PCA(n_components = n_components)
X_train_pca = pca_n2.fit_transform(X_train_std)
X_test_pca = pca_n2.transform(X_test_std)
X_train_pca = pd.DataFrame(X_train_pca, columns=['PC {}'.format(x) for x in range(1, n_components + 1)])
X_test_pca = pd.DataFrame(X_test_pca, columns = ['PC {}'.format(x) for x in range(1, n_components + 1)])


# Create a pairplot of the principal component dataframe
#sns.pairplot(X_train_pca)
#plt.show()

def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
X_train_pca, y_train_pca = create_dataset(X_train_pca, y_train, time_steps = 10)
X_test_pca, y_test_pca = create_dataset(X_test_pca, y_test, time_steps = 10)
```
